```python
# Cria√ß√£o de gr√°fico de dispers√£o com as vari√°veis "AnosNaEmpresa" e "TaxaDi√°ria"

# Generating a scatter plot graph using variables "YearsAtCompany" and "DailyRate"

df1.plot(x="YearsAtCompany", y="DailyRate", kind="scatter", figsize=(10,5), 
         title="YearsAtCompany x DailyRate", color="purple")
```

## Exerc√≠cio 2

**a)** Utilizando a base de dados de covid, crie um gr√°fico de linhas mostrando a evolu√ß√£o do n√∫mero de casos novos m√™s a m√™s, para as DRS "Grande S√£o Paulo", "S√£o Jos√© do Rio Preto" e "Mar√≠lia". Identifique os DRS por cores e legenda. N√£o esque√ßa de colocar t√≠tulo para o gr√°fico e para os eixos. 

```python
# Cria√ß√£o de uma c√≥pia do dataframe original para fazer o pivo.

#Creating a copy of the original dataframe in order to pivot it.

df2_copia = df2.copy()
```

```python
# Pivo do dataframe verificando as vari√°vles "Nome do DRS", "Ano & M√™s" e "Novos Casos" nas cidades solicitadas

# Pivoting the dataframe checking the variables "DRS name", "Year_Month" and "New cases" at the selected cities

df2_pivot = df2_copia.pivot_table(index="ano_mes", columns="nome_drs", values="casos_novos", aggfunc="sum") 
df2_pivot = df2_pivot[["Grande S√£o Paulo", "S√£o Jos√© do Rio Preto", "Mar√≠lia"]]
```

```python
df2_pivot.head()
```

```python
# Cria√ß√£o do gr√°fico para observar as informa√ß√µes coletadas no pivo para as cidades solicitadas

# Creating a graph to observe the information collected through the pivot for the selected cities

ax = df2_pivot.plot(title="Casos de Covid", xlabel="Ano e m√™s", ylabel="N√∫mero de novos casos", figsize=(10,5))

ax.legend(title="Dpto. Regional de Sa√∫de")
```

**b)** Repita o gr√°fico do item a trocando as cores dos DRS.

```python
# Substituindo as cores padronizadas do gr√°fico

# Repeating the above graph replacing its colours

ax_colour = df2_pivot.plot(title="Casos de Covid", xlabel="Ano e m√™s", ylabel="N√∫mero de novos casos", 
                           figsize=(10,5), color={"Grande S√£o Paulo" : "purple", "S√£o Jos√© do Rio Preto" : "teal", 
                                                  "Mar√≠lia" : "pink"})
ax_colour.legend(title="Dpto. Regional de Sa√∫de")
```

**c)** Considerando apenas os DRS 'Bauru', 'S√£o Jos√© do Rio Preto' e 'Mar√≠lia', fa√ßa tr√™s gr√°ficos de pizza, para os meses de abril, maio e junho, mostrando a porcentagem de novos √≥bitos em cada DRS.

üáßüá∑
**Observa√ß√£o**: Fazendo outro pivo com os dados dos novos DRS solicitados e medindo os novos √≥bitos para essas regi√µes.
Como o dataframe √© pequeno, podemos ver que o √≠ndice para o m√™s de "Abril" √© 2, j√° o m√™s de "Junho" √© a √∫ltima posi√ß√£o do dataframe.

A solicita√ß√£o pede uma an√°lise dos meses de Abril at√© Junho, por isso estamos usando o m√©todo de fatiamento  para selecionar as posi√ß√µes que queremos (do √≠ndice 2 at√© o final)

-----------------------------
üá¨üáß
**Note**: Creating another pivot with data of the new requested DRS, measuring new deaths for those cities.
Since the dataframe is small, it's possible to see that the index of "April" is 2, while "June" is in the last position at the dataframe.

The question requested an analysis of April to June; therefore, slicing is being used to select the wanted positions (from index 2 until the end).

```python
drs_bau_sjrp_mar = df2_copia.pivot_table(index="ano_mes", columns= "nome_drs", values="obitos_novos", aggfunc="sum")
drs_bau_sjrp_mar = drs_bau_sjrp_mar[2:][["Bauru", "S√£o Jos√© do Rio Preto", "Mar√≠lia"]]
```

```python
# Cria√ß√£o do gr√°fico de pizza para observar as informa√ß√µes coletadas no pivo para as cidades solicitadas

# Creating a pie graph to observe the information collected through the pivot for the selected cities

ax = drs_bau_sjrp_mar.plot(subplots=True, kind="pie", figsize=(20,15), colors={"#04D8B2", "#800080", "#FFC0CB"}, 
                           autopct="%1.0f%%", labeldistance=None, fontsize=20, title="Novos √≥bitos por Covid-19")
```
